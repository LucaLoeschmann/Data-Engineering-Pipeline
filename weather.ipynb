```python
import pandas as pd
from datetime import datetime
from pytz import timezone
import requests

# Function to fetch weather data for multiple cities
def fetch_weather_data(cities_from_sql, weather_API):
    # Initialize an empty list to store weather data for all cities
    weather_items = []
    
    # Let's ensure the timezone is our local timezone (Berlin in this case)
    berlin_timezone = timezone('Europe/Berlin')
    
    for index, city_row in cities_from_sql.iterrows():
        city = city_row['city_name']
        latitude = city_row['latitude']
        longitude = city_row['longitude']
        
        # weather api
        weather_url = f"https://api.openweathermap.org/data/2.5/forecast?lat={latitude}&lon={longitude}&appid={weather_API}&units=metric"
        weather_response = requests.get(weather_url)
        
        # Check if the API call was successful
        if weather_response.status_code == 200:
            weather_json = weather_response.json()
            
            # Added the time retrieved so we know when the forecast was made
            retrieval_time = datetime.now(berlin_timezone).strftime("%Y-%m-%d %H:%M:%S")
            
            # Iterate through forecast items and extract relevant data
            for item in weather_json["list"]:
                weather_item = {
                    "city": city,
                    "forecast_time": item.get("dt_txt", None),
                    "temperature": item["main"].get("temp", None),
                    "feels_like":item["main"].get("feels_like", None),
                    "forecast": item["weather"][0].get("main", None),
                    "rain_in_last_3h": item.get("rain", {}).get("3h", 0),
                    "wind_speed": item["wind"].get("speed", None),
                    "data_retrieved_at": retrieval_time
                }
                weather_items.append(weather_item)
        else:
            print(f"Error fetching weather data for {city}. Status code: {weather_response.status_code}")
    
    # Convert collected data into a pandas DataFrame
    weather_df = pd.DataFrame(weather_items)
    
    return weather_df
```
